```python
import sevenbridges as sbg
import pandas as pd
import pathlib
import os
import shutil

pd.set_option('display.max_columns', None)
```

```python
## Setup API, set global variables

# default config location is ~/.sevenbridges/credentials
api = sbg.Api(config=sbg.Config(profile='cavatica'))

PROJECT_ID='chapmano/pancancer-ecdna'
PROJ_HOME='..'

# Setup the output directory
def makedirs(path):
    if not os.path.exists(path):
        # Create the directory
        os.makedirs(path)

def cd(path):
    '''
    Change directory (cd) to a location in a project.
    path: list of directory names: myproject/parent/subdir/ --> ['parent','subdir']
    '''
    cd = api.files.query(project=PROJECT_ID,names=[path[0]])[0]
    for subdir in path[1:]:
        cd = api.files.query(parent=cd, names=[subdir])[0]
    return cd
# sv_dir_id=cd(['somatic-simple-variants'])
# files = api.files.query(parent=sv_dir_id).all()
```

```python
sv_dir_id='67f39f6b7baf5a5af9c370b1' # pancancer-ecDNA/files/somatic-simple-variants
local_sv_dir = PROJ_HOME+'/data/variants/somatic'

def download_sb_directory(sb_dir_id,local_dest):
    '''
    download all files in CAVATICA directory sb_dir_id to the directory at local_dest.
    '''
    # create local directory
    out_dir = pathlib.Path(local_dest)
    makedirs(out_dir)
    # download all files
    files = api.files.query(parent=sb_dir_id).all()
    for file in files:
        z = pathlib.Path(out_dir,file.name)
        unz = pathlib.Path(out_dir,z.stem)
        try:
            file.download(path=str(z))
        except sbg.LocalFileAlreadyExists:
            print(f'zip file {file.name} already exists')
        #shutil.unpack_archive(filename=z,extract_dir=unz)
        
        #break
    
download_sb_directory(sv_dir_id,local_sv_dir)
```
