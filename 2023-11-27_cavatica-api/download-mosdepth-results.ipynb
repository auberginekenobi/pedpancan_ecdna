```python
import sevenbridges as sbg
import pandas as pd
import pathlib
import os
import shutil

pd.set_option('display.max_columns', None)
```

```python
## Setup API, set global variables

# default config location is ~/.sevenbridges/credentials
api = sbg.Api(config=sbg.Config(profile='cavatica'))

PROJECT_ID='chapmano/pancancer-ecdna'
PROJ_HOME='..'

# Setup the output directory
def makedirs(path):
    if not os.path.exists(path):
        # Create the directory
        os.makedirs(path)
```

```python
md_dir_id='683e333f7baf5a5af9cec7b4'
local_md_dir = PROJ_HOME+'/data/source/mosdepth/CBTN_mosdepth'


def download_sb_directory(sb_dir_id,local_dest,ext=None,verbose=False):
    '''
    download all files in CAVATICA directory sb_dir_id to the directory at local_dest.
    If ext is not none, only download files ending with that extension.
    '''
    # create local directory
    out_dir = pathlib.Path(local_dest)
    makedirs(out_dir)
    # download all files
    files = api.files.query(parent=sb_dir_id).all()
    for file in files:
        z = pathlib.Path(out_dir,file.name)
        unz = pathlib.Path(out_dir,z.stem)
        if ext is not None:
            if not z.name.endswith(ext):
                continue
        try:
            file.download(path=str(z))
        except sbg.LocalFileAlreadyExists:
            print(f'file {file.name} already exists')
        #shutil.unpack_archive(filename=z,extract_dir=unz)
        if verbose:
            print(z)
    if verbose:
        print('Done!')
    
download_sb_directory(md_dir_id,local_md_dir,'mosdepth.summary.txt',verbose=True)
```
